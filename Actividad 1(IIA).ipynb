    1. función en Python que calcule el máximo común divisor de dos números.

```python
#definimos la función que vamos a utilizar 
def mcd(a, b):
	
	while(b > 0):
		mcd = b
		b = a % b
		a = mcd
	return mcd
 
# solicitamos los dos números y sleccionamos cual es el minimo
# y el maximo
num1 = int(input("Introduce el primer numero: "))
num2 = int(input("Introduce el segundo numero: "))
a = max(num1,num2)
b = min(num1,num2)
 
print("El máximo común divisor de ", a," y ", b," es ", mcd(a, b))
```

    2. función en Python que calcule el mínimo común múltiplo de dos números.

```python
#definimos la función que vamos a utilizar 
def mcd(a, b):
	
	while(b > 0):
		mcd = b
		b = a % b
		a = mcd
	return mcd 
    
#Funcion para hallar el Minimo Comun Multiplo
def mcm(num1, num2):
    a = max(num1, num2)
    b = min(num1, num2)
    mcm = (a / mcd(a, b)) * b
    mcm = int(mcm)
    return mcm 
# Pedimos los datos al usuario
num1 = int(input("Introduce el primer numero: "))
num2 = int(input("Introduce el segundo numero: "))
a = max(num1,num2)
b = min(num1,num2)

# Mostramos el resultado en pantalla
print("El mínimo común múltipl entre ",a," y ",b," es: ",mcm(num1, num2))
```

    3. Número de veces que aparece el artículo «el» en el fichero 

```python
#importamos la libreria re  la cual 
#cuenta con funciones para trabajar 
#con expresiones regulares y cadenas.
import re
import string
archivo.seek(0)#movemos nuestro puntero al inicio del documento
frequency = {} #declaramos un diccionario vacío frecuency
texto_string = archivo.read().lower() 
#almacenamos nuestro texto trasformado en minusculas en un string
limite = re.findall(r' el ', texto_string)

#realizamos un ciclo en todo el mundo del limite
#para contar con cada ves que se repite el articulo
for word in limite:
    count = frequency.get(word,0)
    frequency[word] = count + 1
     
frequency_list = frequency.keys()

for words in frequency_list:
    print("El archivo contine",frequency[words],"articulos 'el'")
    
#obtener todas las palabras y su frecuencia
articulo = frequency[words]
```

    4.	Elimina las tildes del fichero y convierta todas las letras a minúscula.

```python
#asingamos a variables las vocales con mayuculas
#las vocales sin mayusculas
a,b = 'áéíóúü','aeiouu'
#El método maketrans () devuelve una tabla de 
#traducción con un mapeo 1 a 1 de un ordinal 
#Unicode a su traducción / reemplazo.
trans = str.maketrans(a,b)
#mostramos el texto con la tranformacion
#de las tildez y en minusculas
parrafos =(texto_string.translate(trans))
print(parrafos)
```

```python
import os
file = open("resultado.txt", "w")
#primera linea
file.write(lineas+" lineas" + os.linesep)
#segunda linea
file.write(str(articulo)+" articulos 'el'"+ os.linesep)
#tercera linea
file.write(parrafos) 
file = open("resultado.txt")
texto = file.read()
print(texto)
file.close()
archivo.close()
```
